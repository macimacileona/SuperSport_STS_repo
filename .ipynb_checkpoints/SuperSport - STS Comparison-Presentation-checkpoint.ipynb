## PREMATCH

```python


# Group by Operator, Sport, Country, and League, then calculate the average Overround for Match Winner
average_overround_winner = df.groupby(['Operator', 'Sport', 'Country', 'League'])['Match Winner'].agg(lambda x: round(np.nanmean(x), 1) if not x.isnull().all() else np.nan).reset_index()

# Group by Operator, Sport, Country, and League, then calculate the average Overround for Over/Under
average_overround_over_under = df.groupby(['Operator', 'Sport', 'Country', 'League'])['Total(Over/Under)'].agg(lambda x: round(np.nanmean(x), 1) if not x.isnull().all() else np.nan).reset_index()

# Group by Operator, Sport, Country, and League, then calculate the average for No. Of Markets
average_overround_markets = df.groupby(['Operator', 'Sport', 'Country', 'League'])['No. Of Market Lines'].agg(lambda x: round(np.nanmean(x)) if not x.isnull().all() else np.nan).astype('Int64').reset_index()

# Merge the DataFrames on the common columns
average_overround = pd.merge(average_overround_winner, average_overround_over_under, on=['Operator', 'Sport', 'Country', 'League'], suffixes=('_Winner', '_Over/Under'))
average_overround = pd.merge(average_overround, average_overround_markets, on=['Operator', 'Sport', 'Country', 'League'])

# Set option to display all rows in the DataFrame
pd.set_option('display.max_rows', None)

# Create a custom sort order for 'Operator' and 'Sport'
custom_order = ['SuperSport', 'STS']
sport_order = ['Football', 'Basketball', 'Tennis']

# Sort the DataFrame
average_overround['Operator'] = pd.Categorical(average_overround['Operator'], custom_order)
average_overround['Sport'] = pd.Categorical(average_overround['Sport'], sport_order)
average_overround = average_overround.sort_values(by=['Operator', 'Sport', 'Country', 'League'])

# Reset index and display the resulting DataFrame
average_overround.reset_index(drop=True, inplace=True)
#average_overround



```
